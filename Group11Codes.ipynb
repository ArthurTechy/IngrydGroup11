```python
'''
The Questions:


1. Write a program to print twin primes less than 1000. # checked
2. Write a program to find out the prime factors of a number. # checked
3. Write a program to calculate the permutation and combination using the formula:
p(n,r) = n!/(n-r)!
c(n,r) = n!/(r! * (n-r)!) = p(n,r)/r! 
where n is the no of permutations/combinations taken at r time
4. Write a program that converts a decimal number to binary number. 
5. Write a program that returns the product of digits of a given number. # checked
6. Write a program function sumPdivisiors that finds the sum of proper divisors of a given number.
# Proper divisors are those numbers which by which the number is divisible except the number itself. 
# e.g. divisors of 10: 1,2,5
7. Write a program that filters odd numbers in a list. Hint: use lambda func
'''


# Qn 1: Write a program to print twin primes less than 1000.

def is_prime2(num):
    """Func to check if a number is prime or not!"""
    # check if number is prime or not
    for i in range(2, num): #2,3,4,5,6,7,8
        if num % i == 0: # if you find a factor, then num is not prime
            return False
    return True

is_prime2(7)



def twin_prime(num):
    """Func to generate twin prime numbers"""
    
    print(f"TWIN PRIMES ARE in the range of {num} are:\n")
    for first_prime in range(2, num):
        second_prime = first_prime + 2 # formula for twin prime
        # 5, 7 are examples of twin primes
        if is_prime2(first_prime) and is_prime2(second_prime):
            print(f"{first_prime} <==> {second_prime}")

twin_prime(1000)


# Qn 2: Write a program to find out the prime factors of a number.

# Factors of a num
def check_primefactors(num):
    """Func to get the prime factors of a number."""
    factors = []
    prime_factors = []
    
    # check factors
    for i in range(2, num+1):
        if num % i == 0:
            factors.append(i) # get factors of number
    
    # check prime factors
    for n in factors:
        if is_prime2(n):
            prime_factors.append(n)
    return prime_factors 

check_primefactors(15)


'''
Qn 3: Write a program to calculate the permutation and combination using the formula:
p(n,r) = n!/(n-r)!
c(n,r) = n!/(r! * (n-r)!) = p(n,r)/r! 
'''


def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

def factorial(r):
    if r == 0:
        return 1
    else:
        return r * factorial(r-1)
    
def perm(n,r):
    return factorial(n) // factorial(n-r)

def comb(n,r):
    return perm(n,r) // factorial(r)
    
    
n = int(input('enter n value '))
r = int(input('enter r value '))

permutation = perm(n,r)
combination = comb(n,r)
factorial_n = factorial(n)
factorial_r = factorial(r)

print(f'n-factorial is: {factorial_n}')
print(f'r-factorial is: {factorial_r}')
print(f'The permutation is: {permutation}')
print(f'The combination is: {combination}')



#Qn 4: Write a program that converts a decimal number to binary number

def decimal_to_binary(decimal_num):
    """
    Convert a decimal number to a binary number.
    """
    binary_num = bin(decimal_num)
    return binary_num

while True:
    try:
        decimal_output = int(input("Enter a decimal number: "))
        binary_number = decimal_to_binary(decimal_output)
        print("Binary Number:", binary_number[2:])
        break  # Exit the loop if the input and conversion are successful
        
    except ValueError:
        print('Error code: Invalid input. Please enter a valid decimal number.')
        


# Qn 5: Write a program that returns the product of digits of a given number.

def product_of_num(num):
    ''' Func to return the product of digits of a given number.'''
    numb = str(num)
    prod = 1
    
    for n in numb:
        prod *= int(n)   # prod = prod * int(n)
    return prod

product_of_num(454)


'''Qn 6: Write a program function sumPdivisiors that finds the sum of proper divisors of a given number.
# Proper divisors are those numbers by which the number is divisible except the number itself. 
# e.g. divisors of 10: 1,2,5
'''

def divisors_and_sum(num):
    """
    Calculate the list of divisors and their sum for a given number.

    Parameters:
    - num (int): The input number for which divisors and sum are to be calculated.

    Returns:
    - tuple: A tuple containing the list of divisors and their sum.
    """
    divisor_list = []   # Initialize an empty list to store divisors
    divisors_sum = 0    # Initialize a variable to store the sum of divisors

    # Iterate through numbers from 1 to num-1
    for i in range(1, num):
        # Check if i is a divisor of num
        if num % i == 0:
            divisor_list.append(i)   # Add i to the list of divisors
            divisors_sum += i        # Add i to the sum of divisors

    return divisor_list, divisors_sum  # Return a tuple containing the list of divisors and their sum


# Calling the function:
list_of_divisors, sum_of_divisors = divisors_and_sum(10)

# Print the results
print("List of Divisors:", list_of_divisors)
print("Sum of Divisors:", sum_of_divisors)



#Qn 7. Write a program that filters odd numbers in a list. Hint: use lambda func
# Qn 7: method 1

num = [1,2,3,4,5]

odd_filter = list(filter(lambda x: x % 2 != 0, num))
print(odd_filter)


# Qn 7: method 2

my_list = list(range(1, 21))

odd_filter = list(filter(lambda x: x % 2 != 0, my_list))
print(odd_filter)
```
